Los archivos tomados por TOMS-EP se encuentran en: https://acdisc.gsfc.nasa.gov/data/s4pa/EarthProbe_TOMS_Level3/TOMSEPL3dtoz.008/

Es conveniente el realizar un test, para ello se crea una carpeta y se accede:

```python
! mkdir test_ozono_nasa/TOMS_EP && cd test_ozono_nasa/TOMS_EP
```

El siguiente comando descarga datos tomandos por TOMS-EP para un año en especifico, se reemplaza "changeMe" por el año que se desea.

```python
! wget --load-cookies ~/.urs_cookies --save-cookies ~/.urs_cookies --keep-session-cookies -r -c -nH -nd -np -A txt "https://acdisc.gsfc.nasa.gov/data/s4pa/EarthProbe_TOMS_Level3/TOMSEPL3dtoz.008/1996/" 
```
